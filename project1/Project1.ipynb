```python
#Project1
#Peers: Khanh Nguyen, Gabriel Vallangca, Marc Hipona, Leo Azucena

import requests
import pandas as pd
from bs4 import BeautifulSoup
# Get html data from url
r = requests.get('https://www.spaceweatherlive.com/en/solar-activity/top-50-solar-flares')
```

```python
root = BeautifulSoup(r.content,'html5lib')
```

```python
# Get table data
table = root.find('table')
```

```python
# Create new Table
new_table = pd.DataFrame(columns=['rank', 'x_class', 'date', 'region', 'start_time', 'max_time', 'end_time', 'movie'], index = range(0,50))

# Put data in table
row_marker = 0
for row in table.find_all('tr'):
    column_marker = 0
    columns = row.find_all('td')
    row_marker += 1
    for column in columns:
        if (row_marker > 1):
            new_table.iat[row_marker-2,column_marker] = column.get_text()
            column_marker += 1
```

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
r = requests.get('http://cdaw.gsfc.nasa.gov/CME_list/radio/waves_type2.html')
```

```python
root = BeautifulSoup(r.content,'html5lib')
```

```python
table1 = root.find('pre')
```

```python
row_marker = 0
#Extract the whole text as a long string
content = table1.get_text()
#Split each line and put them in an array
lines = content.split('\n')
#Remove the first 11 unnecessary lines
for i in range(0,11):
    lines.pop(0);
#Remove the last three useless lines
lines.pop(len(lines)-1)
lines.pop(len(lines)-1)
lines.pop(len(lines)-1)
#Make the table
new_table1 = pd.DataFrame(columns=['start_date', 'start_time', 'end_date', 'end_time', 'start_frequency', 'end_frequency', 'flare_location', 'flare_region','flare_classification', 'cme_date', 'cme_time', 'cme_angle', 'cme_width', 'cme_speed'], index = range(0,len(lines)))
#Iterate through each line
for line in lines:
    #Split every info in that line
    cols = line.split(' ')
    #Remove any empty info in that array
    while '' in cols:
        cols.remove('')
    col_marker = 0
    #Assign info
    while col_marker < 14:
        new_table1.iat[row_marker, col_marker] = cols[col_marker]
        col_marker += 1
    row_marker += 1
```

```python
new_table1
```

```python
#helper method

def isInt(s):
    try: 
        int(s)
        return True
    except ValueError:
        return False
    
#START OF STEP 4

# Replace all instances of '----' with 'NaN' as stated by the prompt
new_table1 = new_table1.replace('----', 'NaN').replace('-----', 'NaN')

#Populate Halo_flare column
new_table1['is_halo'] = new_table1['cme_angle'].map(lambda x: x == 'Halo')

#Replace "Halo" with 'Na'
new_table1 = new_table1.replace('Halo', 'NA')

#populate widthlowerbound column
new_table1['width_lower_bound'] = new_table1['cme_width'].map(lambda x: str(x)[0] == '>')

#remove any non-numeric part of the width column.
new_table1['cme_width'] = new_table1['cme_width'].map(lambda x: x if isInt(x) else "")

for row in range(0,len(lines)):
    tempArr = new_table1.values[row][0].split('/')
    tempYear = tempArr[0]
    day_month = new_table1.values[row][2]
    new_table1.iat[row, 2] = tempYear + '/' + new_table1.values[row][2]
```

```python
# Combine the date column with the time columns to convert to datetime format
new_table1['start_datetime'] = pd.to_datetime(new_table1['start_date'] + ' ' + new_table1['start_time'])
new_table1['end_datetime'] = ""
```

```python
new_table1 = new_table1[['end_date', 'end_time','start_datetime', 'end_datetime', 'start_frequency', 
                       'end_frequency', 'flare_location', 'flare_region','flare_classification', 
                       'cme_date', 'cme_time', 'cme_angle', 'cme_width', 'cme_speed', 'is_halo', 
                       'width_lower_bound' ]]

for row in range(0,len(lines)):
    tempArr = new_table1.values[row][1].split(":")
    
    if (int(tempArr[0]) < 24):
        new_table1.iat[row, 3] = new_table1.values[row][0].replace('/','-') + ' ' + new_table1.values[row][1] + ':00'
    
```

```python
# Drop the unneeded columns
new_table1 = new_table1.drop('end_date', 1)
new_table1 = new_table1.drop('end_time', 1)


new_table1
```

```python
# Get the rows that have X class flare
topFlares = new_table1.loc[new_table1['flare_classification'].str.contains('X')]
topFlares.is_copy = False
# Get rid of X for sorting purposes
topFlares['flare_classification'] = topFlares['flare_classification'].str.lstrip('X')
# Change type to float
topFlares['flare_classification'] = topFlares.flare_classification.astype(float)
# Sort rows by flare class
topFlares = topFlares.sort_values('flare_classification', ascending = False)
# Extract the top 50
topFlares = topFlares.head(50)
# Put back the X in the flare_classification column values
topFlares['flare_classification'] = topFlares.flare_classification.astype(str)
topFlares['flare_classification'] = "X" + topFlares['flare_classification']
topFlares
```
